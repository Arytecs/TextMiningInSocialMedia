## Text Mining in Social Media - Bag of words

**1- Generamos dos diccionarios llamados ids_train e ids_test que contienen los ids de usuarios clasificados por variedad y sexo. Usamos para ello la funci√≥n generar_dicc_ids**

Para ello abrimos el fichero truth.txt, que contiene estos campos en forma de csv separados por los caracteres ':::'.

```python
# OBTENEMOS UN DICCIONARIO (ids_test) CON LOS IDS DE USUARIO POR VARIEDAD Y SEXO

f = open('./pan-ap17-bigdata/test/truth.txt')

contador=0
sexo={'male':0,'female':0}
variedades={'colombia':0,'argentina':0,'spain':0,'venezuela':0,'peru':0,'chile':0,'mexico':0}

ids_test={'colombia':{'male':[],'female':[]},'argentina':{'male':[],'female':[]},'spain':{'male':[],'female':[]}, \
     'venezuela':{'male':[],'female':[]},'peru':{'male':[],'female':[]},'chile':{'male':[],'female':[]}, \
     'mexico':{'male':[],'female':[]}}

#print(len(f.readlines()))

for line in f:
    line=line.rstrip()
    if len(line)==0:
        continue
    lista=line.split(':::')
    
    ids_test[lista[2]][lista[1]].append(lista[0])
    
    sexo[lista[1]] += 1
    variedades[lista[2]] += 1
    contador+=1

print(sexo,variedades,contador)
```

```python
# OBTENEMOS UN DICCIONARIO (ids_train) CON LOS IDS DE USUARIO POR VARIEDAD Y SEXO

f = open('./pan-ap17-bigdata//training/truth.txt')

contador=0
sexo={'male':0,'female':0}
variedades={'colombia':0,'argentina':0,'spain':0,'venezuela':0,'peru':0,'chile':0,'mexico':0}

ids_train={'colombia':{'male':[],'female':[]},'argentina':{'male':[],'female':[]},'spain':{'male':[],'female':[]}, \
     'venezuela':{'male':[],'female':[]},'peru':{'male':[],'female':[]},'chile':{'male':[],'female':[]}, \
     'mexico':{'male':[],'female':[]}}

#print(len(f.readlines()))

for line in f:
    line=line.rstrip()
    if len(line)==0:
        continue
    lista=line.split(':::')
    
    ids_train[lista[2]][lista[1]].append(lista[0])
    
    sexo[lista[1]] += 1
    variedades[lista[2]] += 1
    contador+=1

print(sexo,variedades,contador)
```

**2- Definici√≥n de funciones:**

Se definen varias funciones para obtener de cada usuario sus tuits y sus etiquetas tanto de g√©nero como de variedad

```python
# FUNCI√ìN QUE DEVUELVE UNA LISTA CON LOS TUITS DE UN ID DE USUARIO

def leer_tuits(id,particion):
    if particion == 'train':
        s='./pan-ap17-bigdata//training/' + id + '.xml'
    else:
        s='./pan-ap17-bigdata//test/' + id + '.xml'
    f=open(s) 
    tuits=list()
    for line in f:
        line=line.rstrip()
        if line.find('<document><![CDATA[') != -1:
            line=line[21:-14]
            tuits.append(line)
    return tuits

def leer_tuits_string(id,particion):
    if particion == 'train':
        s='./pan-ap17-bigdata/training/' + id + '.xml'
    else:
        s='./pan-ap17-bigdata/test/' + id + '.xml'
    f=open(s)
    tuits=list()
    for line in f:
        line=line.rstrip()
        if line.find('<document><![CDATA[') != -1:
            line=line[21:-14]
            tuits.append(line)
    return flatstring(tuits)
```

```python
# FUNCI√ìN QUE CONVIERTE UNA list[list] en list

def flatlist(lista):
    flatlist=list()
    for sublista in lista:
        for item in sublista:
            flatlist.append(item)
    return flatlist

flatlist([[1,2],[3],[4,5,6]])
```

```python
def flatstring(lista):
    s=""
    for item in lista:
        s+=item
    return s

```

```python
# FUNCI√ìN QUE DEVUELVE UNA LISTA CON LOS TUITS DE UNA VARIEDAD Y UN SEXO DADOS

def obtener_tuits(ids,variedad,sexo,particion):
    tuits=list()
    for id in ids[variedad][sexo]:
        tuits.append(leer_tuits(id,particion))
    tuits=flatlist(tuits)
    return tuits
     
def obtener_tuits_string(ids,variedad,sexo,particion):
    tuits=list()
    for id in ids[variedad][sexo]:
        tuits.append(leer_tuits_string(id,particion))
    #tuits=flatlist(tuits)
    return tuits
```

```python
variedades_lista = ['argentina', 'chile', 'colombia', 'mexico', 'peru', 'spain', 'venezuela']
sexo_lista = ['female','male']
variedades_lista_numero = {'argentina':1, 'chile':2, 'colombia': 3, 'mexico': 4, 'peru': 5, 'spain': 6, 'venezuela':7}
```

**3. Generamos los Dataframes train y test**

```python
import pandas as pd

x_train = pd.DataFrame(columns = ['tuits','sexo','variedad'])

# cada fila ser√° un usuario del train. En total debe haber 2800 filas.

i=0
for variedad in variedades_lista:
    v=0
    print(variedad)
    for sexo in sexo_lista:
        s=0
        for id in ids_train[variedad][sexo]:
            t = leer_tuits_string(id,'train')
            x_train.loc[i] = [t,sexo,variedad]
            i+=1
            v+=1
            s+=1
        print("\t",sexo,s)
    print(v)
```

**3.Contar las palabras de hombres y mujeres**

```python
from stop_words import get_stop_words
### cuenta las palabras de hombres y mujeres
wordcountHombre = {}
wordcountMujer = {}
c = 0
a =  get_stop_words('es')
for words in x_train["tuits"].str.split():
    if words not in a:
        
        if x_train["sexo"][c]== "female":
  
            for word in words:
                if word not in wordcountMujer:
                    wordcountMujer[word] = 1
                else:
                    wordcountMujer[word] += 1
        
        
        if x_train["sexo"][c]== "male":
            for word in words:
                if word not in wordcountHombre:
                    wordcountHombre[word] = 1
                else:
                    wordcountHombre[word] += 1
    c += 1
```

Ordenamos en orden descendente las palabras m√°s usadas por hombres y mujeres

```python
#wordcountHombre
#ordena las palabars segun las mas repetidas
max_hombre = sorted(wordcountHombre.items(), key=lambda wordcountHombre: (-wordcountHombre[1], wordcountHombre[0]))
max_Mujer = sorted(wordcountMujer.items(), key=lambda wordcountHombre: (-wordcountHombre[1], wordcountHombre[0]))
```

```python
#aqui se puede ver las palabras mas usadas no aportan mucha informacion por lo que usaremos las stop words

max_hombre[:10]
```

Creamos a funci√≥n **listodict** para transformar las palabras en diccionarios.

```python
#transforma las palabras en diccionario
def listodict(lista):
    mx_dict ={}
    for key , value in lista:
      
        mx_dict[key]= value
    return mx_dict

mx_dictH = listodict(max_hombre)
mx_dictM = listodict(max_Mujer)
    
```

Comparar la frecuencia de aparicion de palabras entre hombre y mujeres haciendo un iftdf

```python
iftdfh = {}
for key, value in mx_dictH.items():
    if key in mx_dictM:
        iftdfh[key] = value * (1/mx_dictM[key])
        
iftdfm = {}
for key, value in mx_dictM.items():
    if key in mx_dictH:
        iftdfm[key] = value * (1/mx_dictH[key])
    
        
```

Creamos una bolsa de palabras con las palabras que son al menos el doble de frecuentes en un g√©nero que en otro

```python
#iftdfh
bag_word_h ={}
for key ,value in iftdfh.items():
    if iftdfh[key] > 3.2: ## relacion TFIDF si una palabra es 3.2 veces mas frecuente en hombres que en mujeres guardala
        bag_word_h[key]= mx_dictH[key]
        
bag_word_m ={}
for key ,value in iftdfm.items():
    if iftdfm[key] > 3.2:   ### relacion TFIDF mujer
        bag_word_m[key]= mx_dictM[key]
```

**4. Mostramos las palabras (TFIDF) m√°s usadas por cada g√©nero**

```python
#mostramos las palabras mas frecuentes de hombres que las mujeres no usan tanto
from wordcloud import WordCloud
#max_hombre.iterrows()
import matplotlib.pyplot as plt
nube = WordCloud( width=2500, height=2000).generate_from_frequencies(bag_word_h)#mx_dict) 
plt.figure(1,figsize=(13, 13))
plt.imshow(nube,interpolation="bilinear")
plt.axis("off")
plt.show()
```

**5. Crea una bolsa de palabras de hombres y mujeres**

```python
#se crea una bolsa de palabras con las palabras con las bag of words de hombres y las mujeres
bag_of_words = []

for word in bag_word_h:
    bag_of_words.append((word,'male'))
    
for word in bag_word_m:
    bag_of_words.append((word,'female'))

#transformamos en un dataframe para poder entrenarlo
new = pd.DataFrame(bag_of_words,columns=['word','sexo'])
```

**6. Crear el modelo de predicci√≥n de g√©nero**

```python
from nltk.tokenize import TweetTokenizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from stop_words import get_stop_words
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import numpy as np
from scipy.sparse.csr import csr_matrix #need this if you want to save tfidf_matrix

# se crea una funcion que tokeniza los tweets que contengan palabras mas utilizadas en la bolsa de palan 
tokenizador = TweetTokenizer(reduce_len=False, preserve_case=True, strip_handles=True)
stop_words = get_stop_words('es')

vec = CountVectorizer(tokenizer=tokenizador.tokenize,stop_words=stop_words,min_df=0.001)
#vec = TfidfVectorizer(tokenizer=tokenizador.tokenize,min_df=0.01)

X_train = vec.fit_transform(new.word)
#X_train
```
